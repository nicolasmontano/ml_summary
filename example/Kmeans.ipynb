```python
import matplotlib.pyplot as plt
from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
import random
```

```python
# Generate sample data
n_samples = 4000
n_components = 5
n_features= 2
random_state=42

X, y = make_blobs(centers=n_components, n_samples=n_samples, n_features=n_features, shuffle=True, random_state=random_state)

# Calculate seeds from k-means++
centers_init, indices = kmeans_plusplus(X, n_clusters=n_components, random_state=random_state)

# Plot init seeds along side sample data
plt.figure(1)
colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m","r"]

for k, col in enumerate(colors):
    cluster_data = y == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1], c=col, marker=".", s=10)

plt.scatter(centers_init[:, 0], centers_init[:, 1], c="b", s=50)
plt.title("K-Means++ Initialization")
plt.xticks([])
plt.yticks([])
plt.show()
```
